# Calculating the variance

```python
#Calculate the mean
mean=df['job'].mean()
```

```python
#Calculate the difference between the value with mean.you will get the distance
difference=df['job']-mean
```

```python
#square the distance
squared_diff=difference**2
```

```python
#calculate the sum of variance obtained
variance=sum(squared_diff)
```

```python
#Calculate the average of variance
variance=variance/len(df['job'])
```
